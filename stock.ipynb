```python
import yfinance as yf
import matplotlib.pyplot as plt

# 株価データを取得する銘柄のティッカーシンボルを指定
ticker_symbol = "AAPL"  # 例: Apple Inc.

# データの取得期間を指定
start_date = "2023-01-01"
end_date = "2023-10-15"

# 株価データを取得
stock_data = yf.download(ticker_symbol, start=start_date, end=end_date)

# 株価データの確認
print(stock_data.head())

# 終値の時間変化をグラフにプロット
plt.figure(figsize=(10, 6))
plt.plot(stock_data['Close'], label='Close Price')
plt.title(f'{ticker_symbol} Stock Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 株価データを取得する銘柄のティッカーシンボルを指定
ticker_symbol = "AAPL"  # 例: Apple Inc.

# データの取得期間を指定
start_date = "2023-01-01"
end_date = "2023-10-15"

# 株価データを取得
stock_data = yf.download(ticker_symbol, start=start_date, end=end_date)

# 株価データの確認
print(stock_data.head())

# グラフの設定
plt.figure(figsize=(12, 8))

# 各データをプロット
plt.plot(stock_data['Close'], label='Close Price', color='blue', linewidth=2)
plt.plot(stock_data['High'], label='High Price', color='green', linestyle='--')
plt.plot(stock_data['Low'], label='Low Price', color='red', linestyle='--')
plt.plot(stock_data['Open'], label='Open Price', color='orange', linestyle=':')

# グラフのタイトルとラベルを設定
plt.title(f'{ticker_symbol} Stock Price Comparison (Open, High, Low, Close)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 比較する銘柄のティッカーシンボルをリストで指定
ticker_symbols = ["AAPL", "MSFT", "GOOGL"]

# データの取得期間を指定
start_date = "2023-01-01"
end_date = "2023-10-15"

# グラフの設定
plt.figure(figsize=(14, 8))

# 各銘柄の株価データを取得し、グラフにプロット
for ticker in ticker_symbols:
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    plt.plot(stock_data['Close'], label=f'{ticker} Close', linewidth=2)
    plt.plot(stock_data['High'], label=f'{ticker} High', linestyle='--')
    plt.plot(stock_data['Low'], label=f'{ticker} Low', linestyle='--')
    plt.plot(stock_data['Open'], label=f'{ticker} Open', linestyle=':')

# グラフのタイトルとラベルを設定
plt.title('Stock Price Comparison (Open, High, Low, Close)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# 比較する銘柄のティッカーシンボルをリストで指定
ticker_symbols = ["AAPL", "MSFT", "GOOGL"]

# データの取得期間を指定
start_date = "2023-01-01"
end_date = "2023-10-15"

# 各銘柄ごとにグラフを作成
for ticker in ticker_symbols:
    # 株価データを取得
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    
    # グラフの設定
    plt.figure(figsize=(12, 6))
    
    # 各データをプロット
    plt.plot(stock_data['Close'], label='Close Price', color='blue', linewidth=2)
    plt.plot(stock_data['High'], label='High Price', color='green', linestyle='--')
    plt.plot(stock_data['Low'], label='Low Price', color='red', linestyle='--')
    plt.plot(stock_data['Open'], label='Open Price', color='orange', linestyle=':')
    
    # グラフのタイトルとラベルを設定
    plt.title(f'{ticker} Stock Price (Open, High, Low, Close)')
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid(True)
    
    # グラフを表示
    plt.show()
```
