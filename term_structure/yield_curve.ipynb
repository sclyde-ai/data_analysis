```python
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイルのURL
url = "https://home.treasury.gov/resource-center/data-chart-center/interest-rates/Daily_Treasury_Bill_Rates.csv"

# CSVデータの読み込み
data = pd.read_csv(url)

# データの確認
print(data.head())

# 日付をdatetime型に変換
data['Date'] = pd.to_datetime(data['Date'])

# グラフの描画
plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['4 Wk Bank Discount Rate'], label='4 Week T-Bill Rate')
plt.plot(data['Date'], data['8 Wk Bank Discount Rate'], label='8 Week T-Bill Rate')
plt.plot(data['Date'], data['13 Wk Bank Discount Rate'], label='13 Week T-Bill Rate')
plt.plot(data['Date'], data['26 Wk Bank Discount Rate'], label='26 Week T-Bill Rate')
plt.plot(data['Date'], data['52 Wk Bank Discount Rate'], label='52 Week T-Bill Rate')

# グラフの装飾
plt.title('US Treasury Bill Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Bank Discount Rate (%)')
plt.legend()
plt.grid()
plt.show()
```
