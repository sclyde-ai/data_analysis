```python
# 国家のリスト
countries = ["Japan", "U.S"]

# 産業分類のリスト（ISIC大分類を簡略化）
isic_sections = [
    "A - 農業、林業及び漁業",
    "B - 鉱業及び採石業",
    "C - 製造業",
    "D - 電気・ガス・蒸気及び空調供給業",
    "E - 水道業、下水処理業、廃棄物処理業",
    "F - 建設業",
    "G - 卸売業及び小売業",
    "H - 運輸業及び倉庫業",
    "I - 宿泊業及び飲食サービス業",
    "J - 情報通信業",
    "K - 金融業及び保険業",
    "L - 不動産業",
    "M - 専門的、科学的及び技術的サービス業",
    "N - 事業施設管理及び事業支援サービス業",
    "O - 公務及び防衛",
    "P - 教育",
    "Q - 保健衛生及び社会事業",
    "R - 芸術、娯楽及びレクリエーション",
    "S - その他のサービス業",
    "T - 家庭内生産活動",
    "U - 国際機関及び外国公館"
]

# 国家をキー、産業分類を値にした辞書を作成
country_industries = {}
for country in countries:
    country_industries[country] = isic_sections

# 結果を表示
for country, industry_list in country_industries.items():
    print(f"{country}: {', '.join(industry_list)}")

import os

# フォルダを作成するディレクトリを指定（オプション）
# 指定しない場合はカレントディレクトリに作成
target_dir = "./stock/"

# 親フォルダとサブフォルダを作成するループ
for parent_folder, subfolders in country_industries.items():
    parent_folder_path = os.path.join(target_dir, parent_folder)
    
    # 親フォルダが存在しない場合のみ作成
    if not os.path.exists(parent_folder_path):
        os.makedirs(parent_folder_path)
        print(f"親フォルダを作成しました: {parent_folder_path}")
    else:
        print(f"親フォルダは既に存在します: {parent_folder_path}")

    # サブフォルダを作成
    for subfolder in subfolders:
        subfolder_path = os.path.join(parent_folder_path, subfolder)
        if not os.path.exists(subfolder_path):  # サブフォルダが存在しない場合のみ作成
            os.makedirs(subfolder_path)
            print(f"サブフォルダを作成しました: {subfolder_path}")
        else:
            print(f"サブフォルダは既に存在します: {subfolder_path}")

print("処理が完了しました。")
```
