```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

# EDINET APIの基本URL
EDINET_API_URL = "https://disclosure.edinet-fsa.go.jp/api/v1"

def get_documents_list(year, month, day, doc_type="2"):
    """
    指定日付の書類一覧を取得
    :param year: 年 (YYYY)
    :param month: 月 (MM)
    :param day: 日 (DD)
    :param doc_type: 書類タイプ (2:有価証券報告書)
    :return: DataFrame
    """
    date_str = f"{year}-{month:02d}-{day:02d}"
    params = {
        "date": date_str,
        "type": doc_type
    }
    
    try:
        response = requests.get(f"{EDINET_API_URL}/documents.json", params=params)
        response.raise_for_status()
        data = response.json()
        
        if data["metadata"]["status"] == "200":
            df = pd.DataFrame(data["results"])
            return df
        else:
            print(f"Error: {data['metadata']['message']}")
            return None
            
    except Exception as e:
        print(f"Error occurred: {e}")
        return None

def download_xbrl_document(doc_id, save_dir="."):
    """
    XBRL文書をダウンロード
    :param doc_id: 書類ID
    :param save_dir: 保存ディレクトリ
    :return: 保存ファイルパス
    """
    try:
        # 書類情報取得
        response = requests.get(f"{EDINET_API_URL}/documents/{doc_id}", headers={"Accept": "application/json"})
        response.raise_for_status()
        doc_info = response.json()
        
        # XBRLファイルのURLを探す
        xbrl_url = None
        for file in doc_info["results"]["attachments"]:
            if file["docType"] == "XBRL" and file["docFileName"].endswith(".zip"):
                xbrl_url = file["docUrl"]
                break
        
        if not xbrl_url:
            print("XBRL file not found")
            return None
        
        # ファイルダウンロード
        file_name = f"{doc_id}_xbrl.zip"
        file_path = f"{save_dir}/{file_name}"
        
        with requests.get(xbrl_url, stream=True) as r:
            r.raise_for_status()
            with open(file_path, 'wb') as f:
                for chunk in r.iter_content(chunk_size=8192):
                    f.write(chunk)
        
        return file_path
        
    except Exception as e:
        print(f"Error occurred: {e}")
        return None

# 使用例
if __name__ == "__main__":
    # 1. 特定の日付の有価証券報告書一覧を取得
    df_docs = get_documents_list(2023, 3, 31)
    if df_docs is not None:
        print(f"Found {len(df_docs)} documents")
        print(df_docs[["docID", "edinetCode", "filerName", "secCode"]].head())
        
        # 2. 最初の企業のXBRLデータをダウンロード
        first_doc = df_docs.iloc[0]
        print(f"\nDownloading XBRL for {first_doc['filerName']}...")
        saved_file = download_xbrl_document(first_doc["docID"])
        if saved_file:
            print(f"File saved to: {saved_file}")
    
    # 適度な間隔を空ける（APIへの負荷軽減のため）
    time.sleep(1)
```
