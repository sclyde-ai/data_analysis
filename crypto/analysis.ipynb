```python
import yfinance as yf

# 暗号通貨のティッカーシンボルを指定
# 通常、通貨ペアは「BTC-USD」のような形式で指定します
ticker = "BTC-USD"  # ビットコインのUSD価格

# データを取得
data = yf.Ticker(ticker)

# 過去の価格データを取得（ここでは過去1ヶ月間のデータ）
hist = data.history()
hist.to_csv(f'data/{ticker}.csv')
print(hist)
```

```python
hist[['Open', 'High', 'Low', 'Close']].plot()
```
